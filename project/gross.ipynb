```python
import numpy as np
import pandas as pd
import seaborn as sb 
import matplotlib.pyplot as plt
```

```python
gd = pd.read_csv('gross-enrollment-ratio-2013-2016.csv',index_col='State_UT')
```

```python
gd
```

```python
gd.replace(to_replace='NR',value=0,inplace=True)
```

```python
gd.Primary_Boys = gd.Primary_Boys.astype(float)
gd.Secondary_Boys = gd.Secondary_Boys.astype(float)
gd.Upper_Primary_Boys = gd.Upper_Primary_Boys.astype(float)
```

```python
gd[['Primary_Boys','Secondary_Boys','Upper_Primary_Boys']]
```

```python
boys = gd['Primary_Boys']+gd['Upper_Primary_Boys']+gd['Secondary_Boys']
```

```python
boys
```

```python
gd.Upper_Primary_Girls = gd.Upper_Primary_Girls.astype(float)
gd.Primary_Girls=gd.Primary_Girls.astype(float)
gd.Secondary_Girls=gd.Secondary_Girls.astype(float)
```

```python
gd[['Upper_Primary_Girls','Primary_Girls','Secondary_Girls']]
```

```python
girls = gd['Primary_Girls']+gd['Secondary_Girls']+gd['Upper_Primary_Girls']
```

```python
girls
```

```python
gd.groupby('State_UT').mean()
```

```python
fig,ax = plt.subplots(1,1,figsize=(15,5))
boys.plot(kind='line',ax=ax)
girls.plot(kind='line',ax=ax)
plt.title('BOYS AND GIRLS RATIO VS STATES ')
plt.xlabel('STATES',size=20)
plt.ylabel('BOYS AND POPULATION')
plt.legend(boys,girls)
plt.show()
```
